# Correlation Heatmap

A correlation matrix is a matrix showing all the correlation coeffiients between the sets of variables. Larger positive values indicate a strong positive relationship, whereas larger negative values indicate a strong negative relationship.

## Meanings

* A correlation coefficient of 1 means that for every positivie increase in one variable, there is a positive increase of fixed proportion in the other. 

* A correlation coefficient of -1 means that for every positive increase in one variable, there is a negative decrease of a fixed proportion in the other. 

* Zero means that there is no relationship between the values. 

```python
# Correlation Matrix Heatmap
f, ax = plt.subplots(figsize=(10,6))
corr = df.corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm", fmt='.2f',linewidths=0.05)
f.subplots_adjust(top=0.9)
t = f.suptitle('Melting point correlation heatmap', fontsize=14)
```
